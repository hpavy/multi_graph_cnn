```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
results = {}
for prob in [0.4, 0.5, 0.6, 0.7, 0.9, 1.]:
    results[prob] = np.load(f"results/proba_{prob}/loss.npz")["loss_rmse"]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (replace with your actual loss values)

plt.figure(figsize=(12, 7))  # Larger figure for better readability

# Plot the loss curve
for prob in results:
    plt.plot(np.arange(len(results[prob]))*20, results[prob], label=f'{prob}', linewidth=2.5, alpha=0.85)



# Logarithmic y-axis
plt.yscale('log')

# Add grid (for both major and minor ticks)
plt.grid(True, which="both", linestyle='--', alpha=0.6)

# Customize axes and labels
plt.xlabel('Iteration/Epoch', fontsize=12, labelpad=10)
plt.ylabel('Loss (log scale)', fontsize=12, labelpad=10)
plt.title('Training Loss Over Time', fontsize=14, pad=12)

# Add legend
plt.legend(fontsize=10, framealpha=1, loc='upper right')

# Customize ticks and spines for a polished look
plt.tick_params(axis='both', which='major', labelsize=10)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)



# Tight layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
plt.savefig("loss.png")
```
